```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
L1 = [1,2,3,4]
L2 = [10,25,32,58]
plt.plot(L1,L2)
plt.title("Simple line chart")
plt.xlabel("Index")
plt.ylabel("Value")
plt.show()
```

```python
L1 = [1,2,3,4]
L2 = [10,25,32,58]
plt.figure(figsize = (8,3)) #resizing is possible only before plt.plot(L1,L2)
plt.plot(L1,L2)
plt.title("Simple line chart")
plt.xlabel("Index")
plt.ylabel("Value")
plt.show()
```

```python
L1 = [1,2,3,4]
L2 = [10,25,32,58]
plt.figure(figsize = (8,3)) #resizing is possible only before plt.plot(L1,L2)
plt.plot(L1,L2, "red")
plt.title("Simple line chart")
plt.xlabel("Index")
plt.ylabel("Value")
plt.show()
```

```python
L1 = [1,2,3,4]
L2 = [10,25,32,58]
plt.figure(figsize = (8,3)) 
plt.plot(L1,L2, "r*")
plt.title("Simple line chart")
plt.xlabel("Index")
plt.ylabel("Value")
plt.show()
```
