**Importing dependancies**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

**A sample line chart**

```python
L1 = [1,2,3,4]
L2 = [10, 25, 32, 58]

#used when a particular variable changes againts the time.

plt.plot(L1, L2)
plt.show()
```

**Adding a title and axis name**

```python
L1 = [1,2,3,4]
L2 = [10, 25, 32, 58]

#used when a particular variable changes againts the time.

plt.plot(L1, L2)
plt.title("Learning line chart")
plt.xlabel("L2 = Time")
plt.ylabel("L1 = Dependant variable")
plt.show()
```

**Changing the Size of the chart**

```python
L1 = [1,2,3,4]
L2 = [10, 25, 32, 58]

#used when a particular variable changes againts the time.
plt.figure(figsize=(6,3))  #figsize should be given before the titile and labeling
plt.plot(L1, L2)
plt.title("Learning line chart")
plt.xlabel("L2 = Time")
plt.ylabel("L1 = Dependant variable")
plt.show()
```

**Change the colors**

```python
L1 = [1,2,3,4]
L2 = [10, 25, 32, 58]

#used when a particular variable changes againts the time.
plt.figure(figsize=(6,3))  #figsize should be given before the titile and labeling
plt.plot(L1, L2, "r*")   #color as third arguement
plt.title("Learning line chart")
plt.xlabel("L2 = Time")
plt.ylabel("L1 = Dependant variable")
plt.show()
```

**Plotting multiple charts together**

```python
x = np.arange(10)
y1 =  x+5
y2 = x**2

plt.figure(figsize=(6,3))  
plt.plot(x, y1, "red",x,y2, "green")   #important : can remove the 2nd "x" and still get the correct output
plt.title("Learning line chart")
plt.xlabel("Time")
plt.ylabel("Dependant variables")
plt.show()
```

# **Subplots** #

# **First method** #

```python
x = np.arange(10)
y1 =  x+5
y2 = x**2
plt.figure(figsize=(6,3)) 

plt.subplot(1,2,1) #1 row 2 column, 1st chart
plt.plot(x, y1, "red")

plt.subplot(1,2,2) #1 row 2 column, 2st chart
plt.plot(x, y2, "green")
plt.show()
```

**2 rows and 1 column**

```python
x = np.arange(10)
y1 =  x+5
y2 = x**2
plt.figure(figsize=(6,3)) 

plt.subplot(2,1,1) #2 row 1 column, 1st chart
plt.plot(x, y1, "red")

plt.subplot(2,1,2) #2 row 1 column, 2st chart
plt.plot(x, y2, "green")
plt.show()
```

# **Second method** #

```python
x = np.arange(10)
y1 =  x+5
y2 = x**2
y3 = -x
y4 = np.sqrt(x)
plt.figure(figsize=(6,3)) 

plt.subplot(2,2,1) #2 row 2 column, 1st chart
plt.plot(x, y1, "red")

plt.subplot(2,2,2) #2 row 2 column, 2nd chart
plt.plot(x, y2, "green")

plt.subplot(2,2,3) #2 row 2 column, 3rd chart
plt.plot(x, y3, "blue")

plt.subplot(2,2,4) #2 row 2 column, 4th chart
plt.plot(x, y4, "yellow")

plt.show()
```

**Another way of doing above task**

```python
x = np.arange(10)
y1 =  x+5
y2 = x**2
y3 = -x
y4 = np.sqrt(x)

fig,sfigs = plt.subplots(nrows = 2, ncols = 2, figsize=(6,3))

sfigs[0,0].plot(x,y1,"r")
sfigs[0,0].set_title("First chart")
sfigs[0,0].set_xlabel("Time")
sfigs[0,0].set_ylabel("Dependant variable")

sfigs[0,1].plot(x,y2,"g")
sfigs[0,1].set_title("Second chart")

sfigs[1,0].plot(x,y3,"k")
sfigs[1,0].set_title("Third chart")

sfigs[1,1].plot(x,y4,"yellow")
sfigs[1,1].set_title("Fourth chart")


plt.show()
```

#  **Area Charts**  #

```python
L1 = [1,2,3,4]
L2 = [10, 25, 32, 58]
plt.figure(figsize=(6,3)) 
plt.fill_between(L1,L2)
plt.show()
```
